# 信頼区間(区間推定)を求める
これによって統計データから誤差と呼べる範囲が分かる

# ライブラリのインポート

```python
import pandas as pd
```

```python
df = pd.read_html("https://bellcurve.jp/statistics/course/8893.html")[0]
df
```

```python
arr = df.values
arr
```

# データの整形

```python
col = arr[0]
val = arr[1:len(arr) - 1]
df = pd.DataFrame(val)
df.columns = col
df
```

```python
import scipy.stats as stats
import numpy as np
```

# 分散が分かっている時

```python
arr = df["全スクリーン数"].astype(int).values
#var=np.var(arr)
var = 5560
mean = np.mean(arr)
sample_num = len(arr)
nrange = stats.norm(loc=mean, scale=np.sqrt(var/sample_num))
low,high = nrange.interval(alpha = 0.95)
print("%.2f<=x<=%.2f"%(low, high))
```

# 分散が分かっていないとき

```python
var = np.var(arr, ddof=1) #不偏分散
trange = stats.t(loc=mean, scale=np.sqrt(var/sample_num), df=sample_num-1)
low, high = trange.interval(alpha = 0.95)
print("%.2f<=x<=%.2f"%(low, high))
```
